```python
import pandas as pd
maps = []
for map in hunter.all_maps:
    if(type(map) == np.ndarray):
        maps.append(map[0].replace(".txt",""))
    else:
        maps.append(map.replace(".txt",""))

sizes = np.array(hunter.total_end_sizes).reshape(len(hunter.total_end_sizes),)
maps = np.array(maps).reshape(len(maps),)

data = pd.DataFrame({"map":maps, "size": sizes})
data_not_dead = data.loc[data["size"]>0]
data_dead = data.loc[data["size"]<0]
print("average size on map, when not dead")
print(data_not_dead.groupby(['map']).mean())
print("-"*40)
print("games on map")
print(data.groupby(['map']).count())
print("-"*40)
print("games when died")
print(data_dead.groupby(['map']).count())
```
